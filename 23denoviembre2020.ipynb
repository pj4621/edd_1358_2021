```
class Nodo:
  def __init__( self , dato):
    self.__dato = dato 
    self.__siguiente = None 

  def get_dato( self ):
    return self.__dato

  def set_dato( self , d ):
    self.__dato = d

  def get_siguiente( self ):
    return self.__siguiente

  def set_siguiente ( self , d ):
    self.__siguiente = d

  #def to_String ( self ):
    #return  "|" + str(self.__dato)+ "|" + str(self.__siguiente)+"|-->" 
  def show( self ):
    curr_node = self # Head
    print(" Inicio de la estructura ligada:")
    print("|" + str(curr_node.get_dato()) + "| -->" , end ="" )
    while ( curr_node.get_siguiente() != None ):
      curr_node = curr_node.get_siguiente()
      print("|" + str(curr_node.get_dato() ) + "| -->" , end ="" )
    print("\nFin")
        

# Ejemplo 1
a = Nodo( 10 )
a.show()

# ejemplo 2
a.set_siguiente( Nodo(20) )
a.show()    

# ejemplo 1 
#a = Nodo (10)
#print(a.to_String())

#ejemplo 2
#a. set_siguiente(20)
#print(a.to_String())
  
# ejemplo 3
a.get_siguiente().set_siguiente(Nodo(30))
a.show() 

# ejemplo 4
a.get_siguiente().get_siguiente().set_siguiente(Nodo(40))
a.show()    

# ejemplo 5
a.get_siguiente().get_siguiente().get_siguiente().set_siguiente(Nodo(50))
a.show() 

# Ejemplo 6 cambio de valor de nodo 1
a.set_dato(12)
a.show()

# Ejemplo 7 cambio de valor de nodo 4
a.get_siguiente().get_siguiente().get_siguiente().set_dato(45)
a.show()

# ejempplo 8 
a.get_siguiente().get_siguiente().get_siguiente().get_siguiente().set_dato(60)
a.show()

# ejemplo 9
a.get_siguiente().set_dato(25)
a.show()
```

```
class Nodo:
    def __init__( self , dato, siguiente=None ):
        self.dato = dato 
        self.siguiente = siguiente 

class LinkedList:
    def __init__( self ):
        self.__head = None 

    def is_empty( self ):
        return self.__head == None
    
    def tail( self ):
        curr_node = self.__head
        while curr_node !=None and curr_node.siguiente != None:
            curr_node = curr_node.siguiente
        return curr_node
    def append ( self , value ):
        nuevo = Nodo ( value )
        if self.__head == None:
            self.__head = nuevo

        else:
            self.tail().siguiente = nuevo
    def transversal( self ):
        curr_node = self.__head
        if curr_node == None:
            print(f"Lista vacia")
        else:
            while curr_node != None:
                print(f"{ curr_node.dato }-->", end= "")
                curr_node = curr_node.siguiente
        print("")
    def prepend( self,value ):
        nuevo = Nodo( value )
        
        if self.__head == None:
            self.__head = nuevo
        
        else:
            x = self.__head
            self.__head = nuevo
            nuevo.siguiente = x

        
```
